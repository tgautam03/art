```python
mod = sio.loadmat('/home/tushar/Dev/py/art/data/03SynMods4.0/Tomo_71_200.mat')
tomo2 = mod['Tomo']
print(tomo2.shape)
```

## True vs Tomo display

```python
numMods = 10
for i in range(numMods):
    k = np.random.randint(tomo.shape[0])
    plt.figure(figsize=(10, 50)) 
    plt.imshow(true[k])
    plt.show()
    plt.figure(figsize=(10, 50)) 
    plt.imshow(tomo[k])
    plt.show()
```
